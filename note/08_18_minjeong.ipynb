```python
m = folium.Map([37.61032322989501,126.91294826169052], zoom_start=12)

folium.Marker(
    location=[37.61032322989501,126.91294826169052],
    tooltip="Click me!",
    popup="Mt. Hood Meadows",
    icon=folium.Icon(icon="cloud"),
).add_to(m)

folium.Marker(
    location=[37.611285900348626,126.91728208529136],
    tooltip="Click me!",
    popup="subway",
    icon=folium.Icon(color="green"),
).add_to(m)

m
```
